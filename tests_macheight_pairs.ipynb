```python
import requests
import json

url = 'https://mach-eight.uc.r.appspot.com/'

response = requests.get(url).json()

sum = 153


class Player:
    
    def __init__(self, dict):
        self.name = dict['last_name']+ ', ' + dict['first_name']
        self.height = int(dict['h_in'])
    


def pairs(arr, sum):
   
    players_dict = {}
    temp_list = []
    second_temp_list = []
    final_temp_list = []
    result_list = []
    seen = set()
   
    
    for ele in arr:
        
        player = Player(ele)
        
        if player.height not in seen:
        
            players_dict[player.height] = [player]
        
        else:
            
            players_dict[player.height].append(player)
        
        seen.add(player.height)
    
    
    for ele in players_dict.keys():
        
        target = sum - ele
        
        if target in players_dict.keys():
            
            temp_list.append(players_dict[ele])
            second_temp_list.append(players_dict[target])
            
    
    aux_list = []
    for ele in temp_list:
        aux_list += ele
            
    
    second_aux_list = []
    for ele in second_temp_list:
        second_aux_list += ele
    
    
    i = 0
    
    while i < len(aux_list):
        zipped_temp_list = zip(aux_list, second_aux_list)
        i += 1
        
    for ele in tuple(zipped_temp_list):
        result_list.append((ele[0].name, ele[1].name))
    
   
    return result_list
```
